```python
#import required libraries

import numpy as np
import pandas as pd


import warnings
warnings.filterwarnings('ignore')
```

```python
# we can drop date column 'dteday' as it is unnecssary( we have year column)
bike_sharing_df.drop(['dteday'],axis=1,inplace= True)
bike_sharing_df.head()
```

```python
#we can also drop 'casual' and 'registered' as our target variable is 'cnt'(casual + registered) and also 'instant'
bike_sharing_df.drop(['casual','registered','instant'],axis=1,inplace= True)
bike_sharing_df.head()
```

```python
#from data we can see that: season,yr,mnth,holiday,weekday,workingday,weathersit all are categorical variables
#We need to replace season,weekday and weathersit with appropriate values as per dictionary
bike_sharing_df['season'].replace({1:'spring', 2:'summer', 3:'fall', 4:'winter'},inplace=True)
bike_sharing_df.head(10)
```

```python
bike_sharing_df['weathersit'].replace({1:'Clear_Few Clouds',2:'Mist_cloudy',3:'Light rain_Light snow_Thunderstorm',4:'Heavy Rain_Ice Pallets_Thunderstorm_Mist'},inplace=True)
bike_sharing_df.head()                                      
                    
```

```python
bike_sharing_df['weekday'].replace({0:"Sunday",1:"Monday",2:"Tuesday",3:"Wednesday",4:"Thursday",5:"Friday",6:"Saturday"},inplace=True)
bike_sharing_df.head()
```

## Step 3: Data preparation

```python
bike_sharing_df.info()
```

```python
#creating dummy variables
season_var = pd.get_dummies(bike_sharing_df['season'],drop_first=True)
month_var = pd.get_dummies(bike_sharing_df['mnth'],drop_first=True)
weather_sit = pd.get_dummies(bike_sharing_df['weathersit'],drop_first=True)
weekday_var = pd.get_dummies(bike_sharing_df['weekday'],drop_first=True)
```

```python
#concat these dummy variable to our dataframe
bike_sharing_df = pd.concat([bike_sharing_df,season_var],axis=1)
bike_sharing_df = pd.concat([bike_sharing_df,month_var],axis=1)
bike_sharing_df = pd.concat([bike_sharing_df,weather_sit],axis=1)
bike_sharing_df = pd.concat([bike_sharing_df,weekday_var],axis=1)
bike_sharing_df.head()
```

```python
#dropping the originals as we have created their dummies
bike_sharing_df.drop(['season'],axis=1, inplace=True)
bike_sharing_df.drop(['mnth'],axis=1, inplace=True)
bike_sharing_df.drop(['weathersit'],axis=1, inplace=True)
bike_sharing_df.drop(['weekday'],axis=1, inplace=True)
```
